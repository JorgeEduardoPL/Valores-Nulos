```python
#Carga desde un archivo .csv sin indice
data = pd.read_csv('Ventas_totales.csv')
data.head(5)
```

```python
#Identificar valores nulos por columna
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data2=data.copy()
```

```python
#Método 1 sustituir valores nulos con promedio o media
data2["salon_ventas"]=data2["salon_ventas"].fillna(round(data["salon_ventas"].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Método 2 sustituir valores nulos con mediana
data2["subtotal_ventas_alimentos_bebidas"]=data2["subtotal_ventas_alimentos_bebidas"].fillna(round(data["subtotal_ventas_alimentos_bebidas"].median(),1))
data2
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Método 3 de sustituir valores nulos por un valor numérico en  concreto
data2["otros_medios"]= data2["otros_medios"].fillna(10)
data2
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Método 4 sustituir valores nulos por un string en  concreto
data2["tarjetas_debito"] =data2["tarjetas_debito"].fillna("5.370589e+07	")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Método 5 sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
data2["tarjetas_credito"] =data2["tarjetas_credito"].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Método 6 de sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
data2["almacen"]=data2["almacen"].fillna(method="bfill")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```
